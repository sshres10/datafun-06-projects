# Section 1-Load: Get the file, store it in your repo, and load it into a DataFrame. 

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diamonds.csv', index_col=0)
```

# Section 2:  Display the first 7 rows and the last 7 rows.

```python
print(df.head(7))
```

# Section 2: Display the last seven rows of the DataFrame.

```python
print(df.tail(7))
```

# Section 3: Descriptive statistics for the numerical columns

```python
print(df[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].describe())
```

# Section 4: Descriptive statistics for the category/text columns

```python
for column in ['cut', 'color', 'clarity']:
    print(df[column].describe())
```

# Section 5:  Print unique category values

```python
for column in ['cut', 'color', 'clarity']:
    print(df[column].unique())
```

# Section 6-Histograms: Use the DataFrame's hist() function to create a histogram for each numerical column.

```python
df.hist(figsize=(10,10))
plt.tight_layout()
plt.show()
```
